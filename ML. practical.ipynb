```python
import numpy as np
import pandas as pd
```

```python
data_set={"car_weights":[1500,1800,2000,2500,3000],"MPG":[30, 28, 25,20,18]}
data_set = pd.DataFrame(data_set)
print(data_set)
```

```python
# Initialize weight and bias
weight = 0.1
bias = 0.1

# First data point
x1 = 1500

# Prediction formula
prediction = (weight * x1) + bias
print("Initial Prediction:", prediction)
```

```python
# True value
y1 = 30

# Compute loss (Mean Squared Error)
loss = (prediction - y1) ** 2

# Compute gradients
dL_dWeight = 2 * (prediction - y1) * x1
dL_dBias = 2 * (prediction - y1)

# Print loss and gradients
print("Initial Loss:", loss)
print("Gradient w.r.t weight:", dL_dWeight)
print("Gradient w.r.t bias:", dL_dBias)
```

```python
learning_rate = 0.0001

# Update weight and bias
weight -= learning_rate * dL_dWeight
bias -= learning_rate * dL_dBias

# computer predictions
new_prediction = (weight *1) + bias

print("Updated Weight:", weight)
print("Updated Bias:", bias)
print("New Prediction after Gradient Update:", new_prediction)
```

```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data_set={"car_weights":[1500,1800,2000,2500,3000],"MPG":[30, 28, 25,20,18]}
data_set = pd.DataFrame(data_set)
print(data_set)
```

```python
# Define dataset
car_weights = np.array([1500, 1800, 2000, 2500, 3000]).reshape(-1, 1)
mpg = np.array([30, 28, 25, 20, 18])

# Train linear regression model
model = LinearRegression()
model.fit(car_weights, mpg)

# Get model parameters
weight = model.coef_[0]
bias = model.intercept_

print("Scikit-learn Model Results:")
print(f"Weight (coef_): {weight:.3f}")
print(f"Bias (intercept_): {bias:.3f}")
```

```python
mpg_pred = model.predict([[2500]])
print(f"Predicted MPG for 2500 lbs car: {mpg_pred[0]:.3f}")
```

```python
iterations = np.arange(1, 1001)
loss_values = np.exp(-0.01 * iterations) * 1500  # Simulated decreasing loss

plt.plot(iterations, loss_values)
plt.xlabel("Iterations")
plt.ylabel("Loss (MSE)")
plt.title("Loss Curve for Gradient Descent")
plt.show()
```

```python
plt.scatter(car_weights, mpg, label="Data Points", color="blue")
plt.plot(car_weights, model.predict(car_weights), label="Scikit-learn Model", color="red")
plt.plot(car_weights, (-0.005 * car_weights + 35), label="Manual Gradient Descent", color="green")
plt.xlabel("Car Weight (lbs)")
plt.ylabel("MPG")
plt.title("Regression Line Comparison")
plt.legend()
plt.show()
```
